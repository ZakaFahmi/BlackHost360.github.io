# Import library yang diperlukan

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from google.colab import files
```

# Unggah file data secara manual

```python

print("Unggah file Excel Anda (misalnya: data_harga_2tahun.xlsx)")
uploaded = files.upload()

# Membaca file Excel
file_name = list(uploaded.keys())[0]  # Mendapatkan nama file
data = pd.read_excel(file_name)

# Menampilkan beberapa baris pertama untuk inspeksi
print(data.head())
```

# Memproses data

```python
# Mengubah kolom 'Bulan' menjadi format numerik (misalnya: nomor bulan)
data['Bulan'] = pd.to_datetime(data['Bulan'])
data['Bulan_Num'] = data['Bulan'].dt.month + (data['Bulan'].dt.year - data['Bulan'].dt.year.min()) * 12

# Mendapatkan daftar unik Kabupaten/Kota
daftar_kabupaten = data['Kabupaten/Kota'].unique()

# Loop melalui setiap Kabupaten/Kota dan visualisasikan hasil prediksi
for kabupaten in daftar_kabupaten:
    # Filter data untuk kabupaten/kota tertentu
    data_kabupaten = data[data['Kabupaten/Kota'] == kabupaten]

    # Fitur dan target
    X = data_kabupaten[['Bulan_Num']]
    y = data_kabupaten['Harga Mean']

    # Membuat model regresi linear
    model = LinearRegression()
    model.fit(X, y)

    # Prediksi harga beras di bulan depan
    bulan_terakhir = data_kabupaten['Bulan_Num'].max()
    bulan_depan = bulan_terakhir + 1
    prediksi_harga = model.predict([[bulan_depan]])

    # Visualisasi hasil
    plt.figure(figsize=(10, 6))
    plt.scatter(X, y, color='blue', label='Data asli')
    plt.plot(X, model.predict(X), color='red', label='Regresi linear')
    plt.scatter([bulan_depan], [prediksi_harga[0]], color='green', label=f'Prediksi Bulan Depan ({prediksi_harga[0]:.2f})', s=100)
    plt.xlabel('Bulan (numerik)')
    plt.ylabel('Harga Mean')
    plt.title(f'Prediksi Harga Beras - {kabupaten}')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Prediksi harga beras di {kabupaten} untuk bulan depan: {prediksi_harga[0]:.2f}")
```
